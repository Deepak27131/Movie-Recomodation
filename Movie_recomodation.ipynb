### Using NLP to convert the data into token or chunk and embedding

```python
df1['tags'] = df1['overview']+' ' + df1['genres']+ " " + df1['tagline']
```

```python
# removing the punctuation because it no meaning in contest and remove stopword (a i the and with or etc) nlp
# and reomoving leminization 

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    text = str(text).lower() # convert all text in lower cas (small letter)
    text = re.sub(r'[^a-zA-Z\s]','',text)  # remove the punctuation in text words
    
    words = text.split() # all word is split into single text 
    
    words = [word for word in words if word not in stop_words] # all word are remove stopword 
    
    # lemmatizes 
    words = [lemmatizer.lemmatize(word)for word in words ]
    
    return " ".join(words)
```

```python
df1['tags'] = df1['tags'].apply(preprocess_text)
```

```python
df1 = df1.reset_index(drop=True)
```

```python
df1.head()
```

```python
indices = pd.Series(df1.index,index = df1['title']).drop_duplicates()
indices
```

```python
## All text converted into vector or embedding(numeric)

from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer

tfidf = TfidfVectorizer(max_features=50000, stop_words='english',ngram_range=(1,2))
```

```python
tfidf_matrix = tfidf.fit_transform(df1['tags'])
```

```python
tfidf_matrix
```

```python
# checking cosine similarity to chek tidf uasing to chek tags feature
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend(title,n = 10):
    if title not in indices:
        return ['Movies not found']
    
    idx = indices[title]
    sim_score = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]
    return df['title'].iloc[similar_idx]
```

```python
recommend('Toy Story')
```

```python
recommend('End Game')
```
